Matplotlib is a widely-used Python library for creating static, interactive, and animated visualizations in a variety of formats, such as graphs, plots, charts, and more. It is a powerful tool for data visualization and is used extensively in data analysis, scientific research, and various fields to convey insights from data to users in a visual form.



** Five common plots that can be created using the Pyplot module of Matplotlib:

1.Line Plot: A basic plot that displays data points connected by straight lines. Useful for showing trends or relationships over time or continuous data.

2.Bar Plot: Displays data using rectangular bars to represent the values. Suitable for comparing categorical data or displaying frequency counts.

3.Scatter Plot: Represents individual data points using dots on a two-dimensional plane. Useful for exploring relationships between two continuous variables.

4.Histogram: Displays the distribution of data by dividing the range of values into intervals (bins) and showing the frequency of data points in each bin.

5.Pie Chart: Represents data as slices of a circle, with each slice proportional to the value it represents. Useful for showing parts of a whole or percentages.

-- A scatter plot is a type of data visualization that uses points (or dots) to represent individual data points on a two-dimensional coordinate system. Each point on the plot corresponds to the values of two variables, where one variable is plotted along the x-axis and the other variable is plotted along the y-axis. Scatter plots are commonly used to visually explore the relationship between two continuous variables and to identify patterns or trends within the data.





```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')

plt.scatter(x,y)
```

The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots in rows and columns, making it easier to visualize and compare different data sets or visualizations.

```python
import numpy as np

import matplotlib.pylab as plt





x= np.array([0,1,2,3,4,5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2,2,1)
plt.plot(x,y1,label = 'line1')
plt.title('plot line1')
plt.legend()


plt.subplot(2,2,2)
plt.plot(x,y2,label = 'line2')
plt.title('plot line2')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y3,label = 'line3')
plt.title('plot line3')
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y4,label = 'line4')
plt.title('plot line4')
plt.legend()


```

A bar plot (also known as a bar chart or bar graph) is a type of data visualization that uses rectangular bars to represent and compare different categories or groups of data. It's particularly useful for visualizing categorical data and comparing the values of different categories, showing the relative sizes or frequencies of those categories.


**Uses of bar plots:

1.Comparative Analysis: Bar plots are commonly used to compare values between different categories, making it easy to see which categories have larger or smaller values.

2.Frequency Distribution: Bar plots can display the frequency distribution of categorical data, showing how often each category appears.

3.Data Distribution: They can be used to show how data is distributed across categories.

4.Change over Time: Bar plots can show changes in categorical data over time or across different conditions.

5.Comparing Subgroups: Grouped or stacked bar plots can help compare subgroups within each category.

6.Presentation: Bar plots are visually intuitive and can be easily understood by a wide audience, making them useful for presentations and reports.

```python
import numpy as np

import matplotlib.pyplot as plt


company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.bar(company, profit)
```

A box plot (also known as a box-and-whisker plot) is a type of data visualization that provides a summary of the distribution of a dataset. It displays the distribution of data across quartiles, allowing you to visualize the spread, central tendency, and presence of outliers in the data. Box plots are particularly useful for understanding the distribution and variability of continuous data.


** Uses of box plots:

1.Distribution Summary: Box plots provide a visual summary of the distribution of data, including the median, spread, and presence of outliers.

2.Comparison: Box plots are useful for comparing distributions across different categories or groups, allowing you to assess similarities and differences in data characteristics.

3.Outlier Detection: Box plots make it easy to identify potential outliers in the data.

4.Skewness and Symmetry: By observing the box's position within the whiskers, you can infer the data's skewness (whether it's skewed to one side) and symmetry.

5.Spread of Data: Box plots give insights into the variability or spread of data within the interquartile range.

6.Understanding Quartiles: Box plots help you understand quartiles and percentiles in the data distribution.

7.Handling Extremes: They can help you understand the distribution when dealing with skewed data or when extremes are important to consider.

8.Data Exploration: Box plots are valuable for initial data exploration, helping you identify potential patterns and trends in the data.

```python
import numpy as np

import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2] , labels = ['A', 'B'])

plt.title('boxplot')
```
